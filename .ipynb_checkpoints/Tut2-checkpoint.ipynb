```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ml_student_age = np.random.randint(18,45,(100))
py_student_age = np.random.randint(15,42,(100))
```

```python
ml_student_age
py_student_age
```

```python
plt.hist(ml_student_age)
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins)
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8)
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, histtype = "step")
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, orientation = "horizontal")
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, orientation = "horizontal", color = "red")
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, orientation = "horizontal", color = "red")
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.legend(["ML Student"])
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, orientation = "vertical", color = "red")
plt.hist(py_student_age, bins, rwidth = 0.8, orientation = "vertical", color = "blue")
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.legend(["ML Student", "Py Student"])
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.8, orientation = "vertical", color = ["red", "blue"])
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.legend(["ML Student", "Py Student"])
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.8, orientation = "vertical", color = ["red", "blue"])
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.legend(["ML Student", "Py Student"])
plt.show()
```

```python
from matplotlib import style

style.use("ggplot")
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.8, orientation = "vertical", color = ["red", "blue"])
plt.title("Machine Learning Age")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students Age")
plt.legend(["ML Student", "Py Student"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10]
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class1_students)
plt.show()
```

```python
plt.barh(classes, class1_students)
plt.show()
```

```python
plt.bar(classes, class1_students, width = 0.2)
plt.show()
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y")
plt.show()
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m", linewidth = 5)
plt.show()
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m", linewidth = 5, alpha = 0.4)
plt.show()
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m", linewidth = 5, alpha = 0.4, linestyle = "--", label = "Class 1 Students", visible = True)
plt.legend()
plt.show()
```

```python
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "y", edgecolor = "m", linewidth = 5, linestyle = "--", label = "Class 1 Students", visible = True)
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "y", edgecolor = "m", linewidth = 5, linestyle = "--", label = "Class 1 Students", visible = True)
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (16, 9))
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "y", edgecolor = "m", linewidth = 5, linestyle = "--", label = "Class 1 Students", visible = True)
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (16, 9))
plt.barh(classes, class1_students, align = "edge", color = "y", edgecolor = "m", linewidth = 5, linestyle = "--", label = "Class 1 Students", visible = True)
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (16, 9))
plt.bar(classes, class1_students, align = "edge", color = "y", edgecolor = "m", linewidth = 5, linestyle = "--", label = "Class 1 Students", visible = True)
plt.title("Bar Chart", fontsize = 18)
plt.xlabel("Classes", fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (16, 9))
plt.bar(classes, class1_students, color = "y", linewidth = 5, linestyle = "--", label = "Class 1 Students", visible = True)
plt.bar(classes, class2_students, color = "b", linewidth = 5, linestyle = "--", label = "Class 2 Students", visible = True)
plt.title("Bar Chart", fontsize = 18)
plt.xlabel("Classes", fontsize = 15)
plt.ylabel("No. of Students", fontsize = 15)
plt.legend()
plt.show()
```
